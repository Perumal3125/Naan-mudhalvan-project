```python
import pandas as pd

try:
    df = pd.read_csv('AirQualityData.csv')
    display(df.head())
except FileNotFoundError:
    print("Error: 'AirQualityData.csv' not found. Please ensure the file is in the correct location or accessible.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")

# Examine the shape of the DataFrame
print("Shape of the DataFrame:", df.shape)

# Investigate data types
print("\nData Types:")
print(df.info())

# Generate descriptive statistics for numerical features
print("\nDescriptive Statistics:")
print(df.describe())

# Check the number of unique values for each column
print("\nNumber of Unique Values:")
for col in df.columns:
    print(f"{col}: {df[col].nunique()}")

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())


# Calculate descriptive statistics (already done)

# Convert 'Date' and 'Time' to datetime objects and analyze trends (already done)
# Remove DateTime index and related code
df = df.reset_index()

daily_avg = df.groupby('Date').mean(numeric_only=True)
print("\nDaily Averages:\n", daily_avg.head())

# Calculate correlation coefficients (already done)

# Analyze 'DayOfWeek' and 'Hour' variations
dayofweek_avg = df.groupby('DayOfWeek').mean(numeric_only=True)
hour_avg = df.groupby('Hour').mean(numeric_only=True)
print("\nAverage Pollutant Levels by Day of Week:\n", dayofweek_avg)
print("\nAverage Pollutant Levels by Hour:\n", hour_avg)


import matplotlib.pyplot as plt

# Histograms
plt.figure(figsize=(15, 10))
pollutants = ['CO(GT)', 'NOx(GT)', 'NO2(GT)', 'O3(GT)', 'SO2(GT)', 'PM2.5', 'PM10']
for i, pollutant in enumerate(pollutants):
    plt.subplot(3, 3, i + 1)
    plt.hist(df[pollutant], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {pollutant}')
    plt.xlabel(pollutant)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# Box plots
plt.figure(figsize=(15, 6))
plt.boxplot([df[col] for col in pollutants], labels=pollutants, patch_artist=True, showfliers=True)
plt.title('Box Plots of Pollutants')
plt.ylabel('Concentration')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Scatter plots
plt.figure(figsize=(10, 6))
plt.scatter(df['PM2.5'], df['PM10'], c=df['Temperature'], cmap='viridis', s=df['WindSpeed'])
plt.title('PM2.5 vs. PM10 (Colored by Temperature, Sized by Wind Speed)')
plt.xlabel('PM2.5')
plt.ylabel('PM10')
plt.colorbar(label='Temperature')
plt.show()

# Line plots for daily average pollutant concentrations
plt.figure(figsize=(15, 12))
for i, pollutant in enumerate(pollutants):
  plt.subplot(3, 3, i + 1)
  plt.plot(df.groupby('Date')[pollutant].mean(), color='orange')
  plt.title(f'Daily Average {pollutant} Over Time')
  plt.xlabel('Date')
  plt.ylabel(f'Average {pollutant}')
  plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Bar charts for average pollutant concentrations by day of week and hour of day
plt.figure(figsize=(15, 10))
for i, pollutant in enumerate(pollutants):
  plt.subplot(3, 3, i + 1)
  df.groupby('DayOfWeek')[pollutant].mean().plot(kind='bar', color='lightgreen')
  plt.title(f'Average {pollutant} by Day of Week')
  plt.xlabel('Day of Week')
  plt.ylabel(f'Average {pollutant}')
  plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 10))
for i, pollutant in enumerate(pollutants):
  plt.subplot(3, 3, i + 1)
  df.groupby('Hour')[pollutant].mean().plot(kind='bar', color='lightcoral')
  plt.title(f'Average {pollutant} by Hour of Day')
  plt.xlabel('Hour of Day')
  plt.ylabel(f'Average {pollutant}')
  plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


import pandas as pd

# Calculate descriptive statistics
numerical_features = df.select_dtypes(include=['number'])
descriptive_stats = numerical_features.describe()
print("Descriptive Statistics:\n", descriptive_stats)

# Convert 'Date' and 'Time' to datetime objects and analyze trends
df['Date'] = pd.to_datetime(df['Date'])
df['DateTime'] = pd.to_datetime(df['Date'].astype(str) + ' ' + df['Time'].astype(str))
df = df.set_index('DateTime')

# Select only numerical columns before resampling and calculating mean
daily_avg = df[numerical_features.columns].resample('D').mean()
print("\nDaily Averages:\n", daily_avg.head())

# Calculate correlation coefficients and visualize in the next step
correlation_matrix = numerical_features.corr()
print("\nCorrelation Matrix:\n", correlation_matrix)

# Analyze 'DayOfWeek' and 'Hour' variations
# Ensure 'DayOfWeek' and 'Hour' are in the DataFrame if needed for these calculations
# Note: 'Hour' column is not explicitly created in the provided code snippets.
# If 'Hour' is needed, it might need to be extracted from the 'DateTime' index.
try:
    dayofweek_avg = df.groupby('DayOfWeek')[numerical_features.columns].mean()
    print("\nAverage Pollutant Levels by Day of Week:\n", dayofweek_avg)
except KeyError:
    print("\n'DayOfWeek' column not found. Skipping calculation of average by Day of Week.")

# Assuming 'Hour' is derived from the index for this calculation
try:
    df['Hour'] = df.index.hour
    hour_avg = df.groupby('Hour')[numerical_features.columns].mean()
    print("\nAverage Pollutant Levels by Hour:\n", hour_avg)
except Exception as e:
    print(f"\nError calculating average by Hour: {e}. Ensure 'Hour' column exists or can be derived.")
```
